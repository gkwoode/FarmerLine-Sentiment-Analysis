```python
from pdfminer.high_level import extract_text

transcript = extract_text("IVRConversations.pdf")
print(transcript)
```

```python
# Remove conversation text
conversations = transcript.split('Conversation ')[1:]
print(conversations)
```

```python
# Replace \n with a space
cleaned_conversations = [conversation.replace('\n', ' ') for conversation in conversations]
print(cleaned_conversations)
```

```python
# Replace double spacing with a single space
texts = [cleaned_conversation.replace('  ', ' ') for cleaned_conversation in cleaned_conversations]
print(texts)
```

```python
# Initialize text_data list
text_data = []

# global dialogue ID
global_dialogue_id = 1

# Iterate over each conversation
for text in texts:
    lines = text.strip().split('●')
    
    # Extract conversation ID and topic
    text_info = lines[0].split(':')
    text_id = int(text_info[0].strip().split()[-1])
    topic = text_info[1].strip()
    
    # Variables to store farmer's and agent's utterances
    farmer_utterances = []
    agent_utterances = []
    
    # Dialogue ID for the current conversation
    dialogue_id = global_dialogue_id
    
    # Iterate over each line in the conversation
    for line in lines[1:]:
        parts = line.split(':')
        if len(parts) >= 2:
            role = parts[0].strip()
            utterance = parts[1].strip()
            
            # Determine the speaker and append utterance accordingly
            if 'Farmer' in role:
                farmer_utterances.append(utterance.strip())
            elif 'Agent' in role:
                agent_utterances.append(utterance.strip())
                
                # Append conversation data to text_data
                text_data.append({
                    'ConversationID': text_id,
                    'DialogueID': dialogue_id,
                    'Topic': topic,
                    'Farmer Utterance': farmer_utterances[-1],
                    'Agent Utterance': agent_utterances[-1]
                })
                
                dialogue_id += 1
                global_dialogue_id += 1

print(text_data)
```
